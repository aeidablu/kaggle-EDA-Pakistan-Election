```python
def wrangle(filepath):
    
    df = pd.read_csv(filepath,encoding = "ISO-8859-1")
    
    #Convert type from object to float
    df['Daily Oil Consumption (Barrels)'] = df['Daily Oil Consumption (Barrels)'].str.replace(",", "")
    df['Daily Oil Consumption (Barrels)'] = df['Daily Oil Consumption (Barrels)'].astype(float)
    df['GDP Per Capita ( USD )'] = df['GDP Per Capita ( USD )'].str.replace(",", "")
    df['GDP Per Capita ( USD )'] = df['GDP Per Capita ( USD )'].astype(float)
    df['Gallons GDP Per Capita Can Buy'] = df['Gallons GDP Per Capita Can Buy'].str.replace(",", "")
    df['Gallons GDP Per Capita Can Buy'] = df['Gallons GDP Per Capita Can Buy'].astype(float)
    
    #Convert World share type from Object to float
    df[['World Share1','World Share2']] = df['World Share'].str.split("%", expand=True)
    df.drop(columns = 'World Share2', inplace=True)
    df['World Share1'] = df['World Share1'].astype(float)
    
    # Drop S# column
    df.drop(columns = 'S#', inplace=True)
    
    return df
```

```python
df = wrangle('../input/petrolgas-prices-worldwide/Petrol Dataset June 23 2022 -- Version 2.csv')
print(df.info())
df.head()
```

# <b> Top ten countries with highest petrol/gas price in usd </b>

```python
df_price_per_liter = df.sort_values('Price Per Liter (USD)', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(12,7))
sns.barplot(
    ax=ax,
    x="Country",
    y= "Price Per Liter (USD)",
    #hue="TURNFEAR",
    data=df_price_per_liter,
    #order=inccat_dict.values()
)
plt.xlabel("Country", fontsize = 15)
plt.ylabel("Price per liter (USD)", fontsize = 15)
plt.title("Top Ten countries with highest Petrol price per liter (USD) by country", fontsize = 15);
```

```python
df_price_per_liter = df.sort_values('Price Per Liter (USD)', ascending=False).tail(10)

fig, ax = plt.subplots(figsize=(12,7))
sns.barplot(
    ax=ax,
    x="Country",
    y= "Price Per Liter (USD)",
    #hue="TURNFEAR",
    data=df_price_per_liter,
    #order=inccat_dict.values()
)
plt.xlabel("Country", fontsize = 15)
plt.ylabel("Price per liter (USD)", fontsize = 15)
plt.title("Top Ten countries with lowest Petrol price per liter (USD) by country", fontsize = 15);
```
