```python
mask = NA02_13["year"] == 2002
NA02 = NA02_13[mask]
print("NA02 shape:",NA02.shape)
```

```python
mask = NA02_13["year"] == 2013
NA13 = NA02_13[mask]
print("NA13 shape:", NA13.shape)
```

```python
# Calculate the mean of four columns 'Votes', 'Total Votes', 'Turnout' and 'year' for each constituency
NA08_state = NA08.groupby('ConstituencyTitle').mean()
# Convert the Turnout column into dict
TU08 = NA08_state[['Turnout']].round(2).sort_values('Turnout',ascending=False).to_dict()
# Make the list of year
Y08=[2008 for i in range(267)]
# Make the list of constituency
S08=list(TU08['Turnout'].keys())
# Make the list of turnout values
P08=list(TU08['Turnout'].values())

NA13_state = NA13.groupby('ConstituencyTitle').mean()
TU13 = NA13_state[['Turnout']].round(2).sort_values('Turnout',ascending=False).to_dict()
Y13=[2013 for i in range(272)]
S13=list(TU13['Turnout'].keys())
P13=list(TU13['Turnout'].values())

NA18_state = NA18.groupby('ConstituencyTitle').mean()
TU18 = NA18_state[['Turnout']].round(2).sort_values('Turnout',ascending=False).to_dict()
Y18=[2018 for i in range(270)]
S18=list(TU18['Turnout'].keys())
P18=list(TU18['Turnout'].values())

# Merged them into one dict
Data1 = {'Year':Y08+Y13+Y18,'ConstituencyTitle':S08+S13+S18,'Turnout':P08+P13+P18}
#Data2 = {'Year':Y18,'ConstituencyTitle':S18,'Turnout':P18}
```

```python
# Convert Data1 dict into dataframe
df_TU08_18 = pd.DataFrame(data=Data1)
df_TU08_18.info()
```

```python
ax = plt.subplots(figsize=(6, 80))
sns.barplot(x=df_TU08_18.Turnout,y=df_TU08_18.ConstituencyTitle,hue=df_TU08_18.Year)
plt.title('Voter Turnout of each constituency in election 2008 / 2013 and 2018', fontsize = 15)
plt.legend(loc= "lower right")
```

```python
# Subset data to election 2002
mask = NA02_13['year'] == 2002
NA02 = NA02_13[mask]
#NA02.info()

Winner02 = pd.DataFrame([])
NOC_02 = NA02['ConstituencyTitle'].nunique()

for i in range(NOC_02):
    Const = "NA-%d"%(i+1)
    NA2002 = NA02.loc[NA02['ConstituencyTitle']==Const, ['Party','Votes','TotalVotes']]
    if NA2002.empty == False:
        MAX = (NA2002.loc[NA2002['Votes'].idxmax()])
        temp = pd.DataFrame([Const,MAX['Party'],MAX['Votes'],MAX['TotalVotes']]).T
        Winner02 = pd.concat([Winner02,temp])
        
Winner02 = Winner02.rename(columns = {0:'Constituency', 1:'Party', 2:'Votes', 3:'TotalVotes'})
Parties02 = np.unique(Winner02.loc[:,'Party'])

Num = pd.DataFrame([])
for i in range (len(Parties02)):
    temp = pd.DataFrame([Parties02[i], len(Winner02.loc[Winner02['Party'] == Parties02[i]])/len(Winner02) ,len(Winner02.loc[Winner02['Party'] == Parties02[i]])]).T
    Num = pd.concat([Num,temp])

Top_02 = Num.rename(columns = {0: 'Party', 1:'Percentage', 2:'Seats Won'})
Top_02 = Top_02.sort_values(by = 'Seats Won', ascending= False )
```
